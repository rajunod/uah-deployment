```python
from IPython.display import Image
import ipywidgets as widgets
```

```python
# User inputs

# Select Freeze Frost Threshold
# 16,20,24,28,32,36
F_THRES = 16
```

```python
def show_map(p):
    with open(f"../data/freeze_images/images/last_freeze_{F_THRES}_p{p}.png", "rb") as f:
        img_widget.value = f.read()
```

```python
img_widget = widgets.Image()
```

```python
widgets.interact(
    show_map,
    p=widgets.IntSlider(min=10,max=90,step=10,value=50)
)

img_widget
```
