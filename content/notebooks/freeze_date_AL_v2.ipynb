```python
import ipywidgets as widgets
```

```python
img_widget = widgets.Image()
```

```python
def show_map(thres,p):
    with open(f"../data/freeze_images/last_freeze_{thres}_p{p}.png", "rb") as f:
        img_widget.value = f.read()
```

```python
threshold_slider = widgets.Dropdown(
    options=[16,20,24,28,32,36],
    value=16,
    description='(°F):'
)

prob_slider = widgets.IntSlider(
    value=50,
    min=10,
    max=90,
    step=10,
    description='Probability:',
    continuous_update=False
)
```

```python
widgets.interactive_output(
    show_map,
    {
        'thres': threshold_slider,
        'p': prob_slider
    }
)
```

```python
widgets.VBox([
    threshold_slider,
    prob_slider,
    img_widget
])
```
